```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

#### Q1. What is the distribution of age of students and what are your observations from the distribution? 

```python
X = [ 3, 11, 12, 19, 22, 23, 24, 25, 27, 29, 15, 6, 27, 17, 29]
df = pd.DataFrame({
    "Age" : X,
})
df['Age'].hist(bins = 4)
#df.describe()
```

#### (2) Question Formed : According to students, which is the most favourite and Least Favourite subject?

```python
no_of_students = [30, 20, 26, 38, 34]
subject_list = ('French', 'English', 'Maths', 'Geography', 'Science')
y_pos = np.arange(len(no_of_students))

plt.bar(y_pos, no_of_students, color=['cyan', 'red', 'cyan', 'yellow', 'cyan'])
plt.xticks(y_pos, subject_list)
plt.show()
```

#### Answer:
1. Red color shows the Least favourite subject i.e English
2. Yellow color shows the most favourite subject i.e Geography 

#### (3) Question Formed : In which activity student spend most of the time and Are the results of the survey is corresponding to company's claim ?

```python
Category = 'School' , 'Sleep' , 'Playing' , 'Study' , 'T.V' , 'Others'
color = ['#FF5E33','#DDFF33' , '#33FFF0' , '#EC33FF', '#33FFAF', '#4F33FF']
statistics = [6,8,2,4,1,3]
explode = (0,0,0,0,0.4,0)
fig , ax = plt.subplots(figsize = (12,8))
ax.pie(statistics,colors = color, explode = explode , labels = Category , autopct = '%1.1f%%',shadow = True , startangle = 90)
ax.axis("equal")
plt.show()
```

#### Answer:
1. Students spend most of the time in sleep activity i.e 33.3%
2. Students spend least time in T.V 
3. Company's claim is not true as T.V only accounts for 4.2% 

#### (4) A clinal trial measurement of two different groups A and B were reported. Which group is performing better (high values are better)?

```python
A=[0.22, -0.87, -2.39, -1.79, 0.37, -1.54, 1.28, -0.31, -0.74, 1.72, 0.38, -0.17, 
   -1.62, -1.10, 0.30, 0.15, 2.30, 0.19, -0.50, -0.09]
B=[-5.13, -2.15, -3.43, -3.83, 0.50, -3.25, 4.32, 0.63, 5.18,
   -0.43, 7.11, 4.87, -3.10, -5.81, 3.76, 6.31, 2.58, 0.07, 5.76, 3.50]
df = pd.DataFrame(
{
    'A_group':A,
    'B_group':B
})
df.plot(kind = 'box',figsize=(12,8) , grid = True)
df.describe()
```

#### Answer : 
1. B-group will perform better than A-group
2. The variability of B-group is much higher than A-group. Furthermore, The interquartile range for group-B is higher than group-A
